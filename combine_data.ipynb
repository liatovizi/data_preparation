```python
#split name column into first_name and last_name
name_split = df_common_club['name'].str.split()

#function to handle the splitting logic
def split_name(name_parts):
    if len(name_parts) >= 3 and name_parts[0] in ['Dr.', 'Ms.', 'Mrs.', 'Mr.']:
        first_name = name_parts[1]
        last_name = name_parts[2] + ' ' + name_parts[0]
    else:
        first_name = name_parts[0]
        last_name = ' '.join(name_parts[1:])
    return pd.Series([first_name, last_name])
```

```python
df_common_club[['first_name', 'last_name']] = name_split.apply(split_name)
```

```python
#load data into the model

conn = sqlite3.connect('membership.db')

# Create tables
conn.execute('''
CREATE TABLE IF NOT EXISTS Company (
    company_id INTEGER PRIMARY KEY AUTOINCREMENT,
    company_name TEXT NOT NULL
)
''')

conn.execute('''
CREATE TABLE IF NOT EXISTS Member (
    member_id INTEGER PRIMARY KEY AUTOINCREMENT,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    date_of_birth DATE,
    last_active DATE,
    member_since DATE,
    state TEXT,
    data_source TEXT,
    company_id INTEGER,
    FOREIGN KEY (company_id) REFERENCES Company(company_id)
)
''')


```

```python
# Insert data into Company table
for company_name in combined_df['company_name'].unique():
    conn.execute('INSERT INTO Company (company_name) VALUES (?)', (company_name,))

# Insert data into Member table
for _, row in combined_df.iterrows():
    company_id = conn.execute('SELECT company_id FROM Company WHERE company_name = ?', (row['company_name'],)).fetchone()[0]
    conn.execute('''
    INSERT INTO Member (first_name, last_name, date_of_birth, last_active, member_since, state, data_source, company_id)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
    ''', (row['first_name'], row['last_name'], row['date_of_birth'], row['last_active'], row['member_since'], row['state'], row['data_source'], company_id))

# Commit changes and close connection
conn.commit()
conn.close()
```
